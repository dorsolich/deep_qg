### This notebooks:
- reads (import) the reference and generated questions by the basic setting, as well as the questions metrics,
- saves all the info in an excel file called basic_questions_and_metrics.xlsx

```python
import notebook_hook
import pathlib
import json
from tqdm import tqdm
import numpy as np
from qg.results_analysis.objects.POSAnalysis import POS_analysis_object
```

```python
PACKAGE_ROOT = pathlib.Path().resolve().parents[1]
print(f"Root directory: {PACKAGE_ROOT}")
```

### Reading the questions generated by the basic setting, and the metrics scores.

```python
DATA_DIR_AA = PACKAGE_ROOT/'qg'/'transformers_models'/'experiment_AA'
DATA_DIR_AQPL = PACKAGE_ROOT/'qg'/'transformers_models'/'experiment_AQPL'
DATA_DIR_BASIC = PACKAGE_ROOT/'qg'/'transformers_models'/'experiment_basic'
DATA_DIR_OQPL = PACKAGE_ROOT/'qg'/'transformers_models'/'experiment_OQPL'

results_folders = ["AA"]
splits = ["validation"]
pos_results = {}

for split in splits:
    for folder in results_folders:

        # Uploading generated questions...
        with open(PACKAGE_ROOT/f"transformers_models/experiment_{folder}/mapped_{split}_questions.json", encoding="utf-8") as f:
            questions = json.load(f)
            predictions = questions["predictions"]
            references = questions["references"]

        with open(PACKAGE_ROOT/f"transformers_models/experiment_{folder}/classification_{split}_questions.json", encoding="utf-8") as f:
            questions = json.load(f)
            labels = questions["pred_y"]

        with open(PACKAGE_ROOT/f"transformers_models/experiment_{folder}/scores_{split}_questions.json", encoding="utf-8") as f:
            questions = json.load(f)
            cos_similarity = questions["examples_similarity"]
            bleu1_t = questions["bleu_1_tokens"]
            bleu2_t = questions["bleu_2_tokens"]
            bleu3_t = questions["bleu_3_tokens"]
            bleu4_t = questions["bleu_4_tokens"]
            bleu1_l = questions["bleu_1_lemmas"]
            bleu2_l = questions["bleu_2_lemmas"]
            bleu3_l = questions["bleu_3_lemmas"]
            bleu4_l = questions["bleu_4_lemmas"]
```

### Saving it into a CSV file to analyze it visually

```python
import pandas as pd

df = pd.DataFrame.from_dict({ 
    "references": references, 
    "predictions": predictions, 
    "labels": labels,
    "cosine_similarity": cos_similarity,
    "BLEU1-T": bleu1_t,
    "BLEU2-T": bleu2_t,
    "BLEU3-T": bleu3_t,
    "BLEU4-T": bleu4_t,
    "BLEU1-L": bleu1_l,
    "BLEU2-L": bleu2_l,
    "BLEU3-L": bleu3_l,
    "BLEU4-L": bleu4_l,

    })
df.to_excel("basic_questions_and_metrics_.xlsx")
```
